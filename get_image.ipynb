# For Exporting Needed Images

```python
import json
import shutil
import os
```

```python
# Path to your JSON file
json_file_path = 'asset/local_image_data.json'

# New folder where you want to move the images
new_folder_path = 'upload_image'

# Create the new folder if it does not exist
os.makedirs(new_folder_path, exist_ok=True)

# Read the JSON file
with open(json_file_path, 'r') as file:
    data = json.load(file)
```

```python
data = data['nodes_info']
```

```python
data[5]
```

```python
# Loop over each image, enumerate provides the index
for index, image in enumerate(data):
    # Get the image path from the JSON
    path = image['path']
    image_path = f'asset/{path}'  # Adjust this line if the path needs further processing
    print(image_path)
    # Define the new path for the image in the new folder with the index as part of the file name
    new_image_path = os.path.join(new_folder_path, f"{index}_{os.path.basename(image_path)}")
    # Copy the image to the new folder
    shutil.copy(image_path, new_image_path)
```

```python
# Path to your JSON file
json_file_path = 'asset/local_image_data_new.json'

# New directory where the paths should be updated to
new_directory_path = 'https://raw.githubusercontent.com/KevinBian107/laplacian_eigenmap_vis/master/upload_image/'

# Read the JSON file
with open(json_file_path, 'r') as file:
    data = json.load(file)

data = data['nodes_info']

# Update the image paths in the JSON data
for image in data:
    # Extract the original filename from the path
    original_filename = os.path.basename(image['path'])
    # Update the path with the new directory
    image['path'] = os.path.join(new_directory_path, original_filename)

# Write the updated JSON data back to the file
with open(json_file_path, 'w') as file:
    json.dump(data, file, indent=4)
```
